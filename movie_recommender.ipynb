```python
import pandas as pd
import numpy as np
import ast 
```

```python
# the useful columns for the following system will be
#genres,
#id,
#keywords,
#title
#overview,
#cast
#crew
#movies =  movies[['movie_id','crew','cast','title','overview','keywords','genres']]
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

# PreProcessing


```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python

## the ast.literal_eval is used to convert the string of objects to objects'
def convert_helper(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L

movies['genres'] = movies['genres'].apply(convert_helper)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert_helper)
```

```python

movies.head()
```

```python
## creating a function to store only the top 3  actors
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter+=1
    return L 
```

```python
#movies['cast'] = 
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
movies.head(1)
```

# finding the name of the director

```python
def find_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
            
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

# creation of tags
now we will replace all the space in an identity for example we have Sam Worthington and Sam Mendis
it should be replaced with  SamWorthington and SamMendis
this is done to ensure that the correct tags are been used by the recommender system it should niot get confused with Sam of Worthington and Mendis


```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())## conversion to a list of words
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['genres']
```

```python
new_movies = movies[['movie_id','tags','title']]
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : " ".join(x))
```

```python
new_movies.head(1)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x :x.lower())
```

```python
new_movies.head()
```

# stemming
since currently the words such as act acts actor actors , activity and activities are all been treated differntly therefore stemming is used which convers ['love','loving','loved'] to ['love','love','love']

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))

    return " ".join(l)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
new_movies['tags']
```

```python
new_movies['tags']
```

# vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer( max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
# tbe above 2d array stores the distance ot each vector with another vector
for i in similarity[0]:
    print(i)
```

```python
list(enumerate(similarity[0]))
# this is used to convert the similarity matrice into the pair of index , distance 
# 0th movie and index ka distance is stored in the tuple
```

```python
def recommend(movie):
    mov_index = new_movies[new_movies['title']==movie].index[0]
    #distances = list(enumerate(similarity[mov_index]))
    distances = similarity[mov_index]
    
    movie_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:6]
    for i in movie_list:
        print (new_movies.iloc[i[0]].title)
```

```python
new_movies[new_movies['title']=='Batman Begins'].index[0]
```

```python
recommend('Spectre')
```
